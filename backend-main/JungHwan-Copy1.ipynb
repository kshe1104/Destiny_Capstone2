```python
import pandas as pd
from haversine import haversine

def find_meeting_points(data_A, data_B):
    meeting_points = []

    for i in range(min(len(data_A), len(data_B))):
        location_A = (data_A['시작 위치 (Latitude)'].values[i], data_A['시작 위치 (Longitude)'].values[i])
        location_B = (data_B['시작 위치 (Latitude)'].values[i], data_B['시작 위치 (Longitude)'].values[i])
        distance = haversine(location_A, location_B, unit='m')

        if distance <= 20:
            time_A_start = data_A['시작 날짜'].values[i]
            time_A_end = data_A['종료 날짜'].values[i]
            time_B_start = data_B['시작 날짜'].values[i]
            time_B_end = data_B['종료 날짜'].values[i]

            if time_A_end < time_B_start:
                meeting_point = f"{location_A}에서 {time_A_end} ~ {time_B_start}에서 만났을 수 있다."
            elif time_B_end < time_A_start:
                meeting_point = f"{location_B}에서 {time_B_end} ~ {time_A_start}에서 만났을 수 있다."
            else:
                meeting_point = f"{location_A}에서 {max(time_A_start, time_B_start)}  ~ {min(time_A_end, time_B_end)}에서 만났을 수 있다."

            meeting_points.append(meeting_point)

    return meeting_points


meeting_points = find_meeting_points(data1, data2)

for point in meeting_points:
    print(point)

```

```python
import os
import uuid
from flask import Flask, render_template, request
from werkzeug.utils import secure_filename
import json

app = Flask(__name__)

# 업로드할 디렉토리 설정
UPLOAD_FOLDER1 = '/home/s60181908/tmp1'
UPLOAD_FOLDER2 = '/home/s60181908/tmp2'
app.config['UPLOAD_FOLDER'] = UPLOAD_FOLDER1

def allowed_file(filename):
    return '.' in filename and filename.rsplit('.', 1)[1].lower() == 'json'

@app.route('/', methods=['GET'])
def index():
    return 'Hello World!'

@app.route('/file_upload', methods=['POST'])
def upload_file():
    if request.method == 'POST':
        # 'file' 필드에서 파일 리스트 가져옴
        files = request.files.getlist('file')

        for idx, file in enumerate(files, start=1):
            if file:
                filename = secure_filename(file.filename)
                if allowed_file(filename):
                    unique_filename = str(uuid.uuid4()) + '_' + filename
                    if idx == 1:
                        file.save(os.path.join(app.config['UPLOAD_FOLDER1'], unique_filename))
                    elif idx == 2:
                        file.save(os.path.join(app.config['UPLOAD_FOLDER2'], unique_filename))
                else:
                    return "Error: json 파일만 업로드 가능합니다."

        return "Files uploaded successfully"

    return '''
    <!doctype html>
    <title>File Upload</title>
    <h1>Upload Files</h1>
    <form method=post enctype=multipart/form-data>
      <input type=file name=file>
      <input type=file name=file>
      <input type=submit value=Upload>
    </form>
    '''

if __name__ == '__main__':
    app.run(host='0.0.0.0')
```

```python
import pandas as pd
from haversine import haversine

def remove_data_by_location(df, latitude, longitude, max_distance=50):
    def is_within_max_distance(row):
        start_coords = (row['시작 위치 (Latitude)'], row['시작 위치 (Longitude)'])
        end_coords = (row['종료 위치 (Latitude)'], row['종료 위치 (Longitude)'])
        
        start_distance = haversine((latitude, longitude), start_coords, unit='m')
        end_distance = haversine((latitude, longitude), end_coords, unit='m')
        
        return start_distance > max_distance and end_distance > max_distance
    
    filtered_df = df[df.apply(is_within_max_distance, axis=1)]
    return filtered_df

filtered_data = remove_data_by_location(df, 37.571972, 126.980377, max_distance=50)
filtered_data

```
